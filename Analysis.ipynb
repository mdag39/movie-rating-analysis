```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('D:\\python project\\dataset.csv',na_values=['nan'])
```

```python
df
```

```python
df=df.drop('aspect_ratio',axis=1)#Data Reduction
```

```python
df=df.drop('movie_imdb_link',axis=1)
df=df.drop('color',axis=1)
```

```python
#dropping null values in 'title year
a=[]
for i in range(len(df)):
    if np.isnan(df['title_year'][i]):
        df=df.drop(i)
```

```python
#deleting rowsbelow year 2000
for i in df.index:
    if df.title_year[i] < 2000:
        df=df.drop([i],axis=0)
```

```python
df=df.drop_duplicates()#data cleaning
```

```python
#haNDLInG ZEROS        
me_fa=np.round(df['facenumber_in_poster'].mean())
for i in df.index:
    if df['facenumber_in_poster'][i]==0:
            df.facenumber_in_poster.loc[i]=(me_fa)#putting round off mean of the series
```

```python
#data Manipulation
df=df.sort_values('imdb_score',ascending=False)
```

```python
#Data Analysis
for i in range(3):
    df['title_year'][i]=int(df['title_year'][i])
min_year=min(df['title_year'])
print(min_year)
max_year=max(df['title_year'])
max_year
```

# PROBLEM 6. Which movies is the top profit and loss.

```python
#handling nan in gross

me_gr=np.round(df['gross'].mean())
for i in df.index:
    if np.isnan(df['gross'][i]):
        df=df.drop(i)
print(len(df))
#handling nan in budget
me_bu=np.round(df['budget'].mean())
for i in df.index:
    if np.isnan(df['budget'][i]):
        df=df.drop(i)
print(len(df))        

```

```python
Profit=df['gross']-df['budget']
```

```python
Profit=pd.DataFrame(Profit)
Profit.columns=['Profit']
```

```python
df=pd.concat([df,Profit],axis=1)
```

```python
#sorting dataframe according  to budget
df=df.sort_values(by=['Profit'],ascending=False)
```

```python
tp_pr=df.head(20)
tp_pr
```

```python
lw_pr=df.tail(20)
lw_pr
```

```python
plt.figure(figsize=(15,5))
plt.bar(tp_pr['movie_title'],tp_pr['Profit'],color='red')
plt.title("Top Profit Movies ",fontsize=20)
plt.xlabel('Movie title',fontsize=20)
plt.ylabel('Profit(in Billions)',fontsize=20)
plt.tick_params(labelsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(lw_pr['movie_title'],-(lw_pr['Profit']),color='red')
plt.title("Top Loss Movies ",fontsize=20)
plt.xlabel('Movies',fontsize=20)
plt.ylabel('Loss(in Billions)',fontsize=20)
plt.xticks(rotation='vertical')
plt.ylim(0,1000000000)
plt.tick_params(labelsize=15)
plt.show()
```

# PROBLEM 6. Which Movie Has The Highest Return On Investment.# return on investment

```python

Roi=pd.DataFrame(df['Profit']/df['budget']*100)
Roi.columns=['roi']
```

```python
df=pd.concat([df,Roi],axis=1)
```

```python
df=df.sort_values(by=['roi'],ascending=False)
tp_roi=df.head(10)
```
